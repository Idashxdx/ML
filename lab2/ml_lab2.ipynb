**2. EDA**

**2.1.** Общая информация

```python
print(f"Размер train.csv: {train.shape}")
print(f"Размер test.csv: {test.shape}")
print("\ntrain.csv:\n")
print(train.info())
print("\ntest.csv:\n")
print(test.info())
```

Все признаки, кроме id (int64), имеют тип float64.

**2.2.** Анализ пропущенных и повторяющихся данных

```python
print("\nПропущенные значения в train.csv:\n")
print(train.isnull().sum())
print("\nПропущенные значения в test.csv:\n")
print(test.isnull().sum())
```

```python
print("Количество повторов в train.csv:", train.duplicated().sum())
print("Количество повторов в test.csv:", test.duplicated().sum())
```

Пропущенные значения и повторы отсутствуют.


**2.3.** Анализ и обработка целевой переменной smoking

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='smoking', data=train)
plt.title('Распределение целевой переменной smoking')
plt.show()

print("\nСтатистики по целевой переменной:")
print(train['smoking'].value_counts(normalize=True))
```
