**2.4.** Анализ и обработка числовых признаков

```python
numeric_columns = train.select_dtypes(include=['float64']).columns.tolist()
numeric_columns = [col for col in numeric_columns if col not in ['id', 'smoking']]
print("Все числовые признаки:", numeric_columns)
```

```python
import numpy as np
```

```python
from scipy.stats import skew

skewness = train[numeric_columns].apply(skew).sort_values(ascending=False)

print(skewness)
plt.figure(figsize=(20, 15))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(6, 5, i)
    sns.histplot(train[col], kde=True)
    plt.title(f'{col}\nSkew: {skewness[col]:.2f}\n')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for col in ['hearing(right)', 'hearing(left)', 'Urine protein', 'AST', 'Gtp', 'eyesight(left)', 'eyesight(right)',
         'fasting blood sugar', 'ALT', 'dental caries', 'triglyceride'
    ]:
    print(f"\nРаспределение для {col}:")
    print(train[col].value_counts())
```

```python
skewed_columns = [
    'AST', 'Gtp', 'fasting blood sugar', 'ALT', 'triglyceride'
]
for col in skewed_columns:
    train[col] = np.log1p(train[col])
    test[col] = np.log1p(test[col])
```

```python
for col in ['eyesight(left)', 'eyesight(right)']:
    train[col] = train[col].replace(9.9, train[col][train[col] != 9.9].max())
    test[col] = test[col].replace(9.9, test[col][test[col] != 9.9].max())
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(6, 5, i)
    sns.boxplot(x='smoking', y=col, data=train)
    plt.title(f'Распределение {col}')
    plt.xlabel('Статус курения')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
display(train[numeric_columns].describe())

outliers_count = {}
for col in numeric_columns:
    Q1 = train[col].quantile(0.05)
    Q3 = train[col].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = train[(train[col] < lower_bound) | (train[col] > upper_bound)]
    outliers_count[col] = len(outliers)

print("\nКоличество выбросов по признакам:")
for col, count in outliers_count.items():
    print(f"{col}: {count} выбросов")
```

```python
train['waist(cm)'] = train['waist(cm)'].replace(0.9, train['waist(cm)'].median())
```

```python
max_reasonable_hb = 20
train['hemoglobin'] = train['hemoglobin'].clip(upper=max_reasonable_hb)
test['hemoglobin'] = test['hemoglobin'].clip(upper=max_reasonable_hb)
```

**2.5.** Матрица корреляций

```python
numeric_data = train[numeric_columns]

plt.figure(figsize=(20, 15))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1,
            annot_kws={"size": 8}, linewidths=0.5)
plt.title("Матрица корреляций", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
correlation_with_target = train.corr()['smoking'].abs().sort_values(ascending=False)
print("Корреляция признаков с целевой переменной (smoking):")
print(correlation_with_target[1:])
```
