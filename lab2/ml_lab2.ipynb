**2.4.** Анализ и обработка числовых признаков

```python
numeric_columns = train.select_dtypes(include=['float64']).columns.tolist()
numeric_columns = [col for col in numeric_columns if col not in ['id', 'smoking']]
print("Все числовые признаки:", numeric_columns)
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(6, 5, i)
    sns.boxplot(x='smoking', y=col, data=train)
    plt.title(f'Распределение {col}')
    plt.xlabel('Статус курения')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

Признаки с явными выбросами:

    waist(cm) - несколько очень высоких значений

    fasting blood sugar, triglyceride - много высоких выбросов

    AST - заметные выбросы вверх

    ALT - значительные выбросы

    Gtp - много экстремальных значений

```python
display(train[numeric_columns].describe())

outliers_count = {}
for col in numeric_columns:
    Q1 = train[col].quantile(0.05)
    Q3 = train[col].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = train[(train[col] < lower_bound) | (train[col] > upper_bound)]
    outliers_count[col] = len(outliers)

print("\nКоличество выбросов по признакам:")
for col, count in outliers_count.items():
    print(f"{col}: {count} выбросов")
```

Наибольшее количество выбросов наблюдается в признаках: hearing(left и right), Urine protein, Gtp, ALT, AST, fasting blood sugar.

Признаки hearing(left) и hearing(right) имеют бинарные или близкие к бинарным значениям (1 и 2), что может быть категориальным признаком.

Признак dental caries также является бинарным (0 или 1).

Некоторые признаки, такие как waist(cm) и hemoglobin, имеют единичные выбросы.

Признаки eyesight(left и right) имеют по 3 выброса, что может быть связано с аномальными значениями зрения.

```python
import numpy as np
```

```python
def winsorize_df(df, columns, lower_quantile=0.05, upper_quantile=0.95):
    df = df.copy()
    for col in columns:
        lower = df[col].quantile(lower_quantile)
        upper = df[col].quantile(upper_quantile)
        df[col] = df[col].clip(lower, upper)
    return df

columns_to_winsorize = [
    'Gtp', 'ALT', 'AST',
    'fasting blood sugar', 'triglyceride',
    'waist(cm)', 'hemoglobin'
]

train = winsorize_df(train, columns_to_winsorize)
test = winsorize_df(test, columns_to_winsorize)

skewed_columns = ['Gtp', 'ALT', 'AST', 'triglyceride']
for col in skewed_columns:
    train[col] = np.log1p(train[col])
    test[col] = np.log1p(test[col])
```

```python
plt.figure(figsize=(15, 15))
for i, col in enumerate(columns_to_winsorize, 1):
    plt.subplot(6, 5, i)
    sns.boxplot(x='smoking', y=col, data=train)
    plt.title(f'Распределение {col}')
    plt.xlabel('Статус курения')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
train['waist(cm)'] = train['waist(cm)'].replace(0.9, train['waist(cm)'].median())
train['eyesight(left)'] = train['eyesight(left)'].clip(upper=2.0)
train['eyesight(right)'] = train['eyesight(right)'].clip(upper=2.0)
for col in ['hearing(left)', 'hearing(right)', 'dental caries']:
    print(f"\nРаспределение для {col}:")
    print(train[col].value_counts())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='age', hue='smoking', data=train)
plt.title('Распределение возраста по статусу курения')
plt.xticks(rotation=45)
plt.show()
```

**2.5.** Матрица корреляций

```python
numeric_data = train[numeric_columns]

plt.figure(figsize=(20, 15))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1,
            annot_kws={"size": 8}, linewidths=0.5)
plt.title("Матрица корреляций", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
correlation_with_target = train.corr()['smoking'].abs().sort_values(ascending=False)
print("Корреляция признаков с целевой переменной (smoking):")
print(correlation_with_target[1:])
```
